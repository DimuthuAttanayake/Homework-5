```python
#importing libraries
import requests
```

```python
#3.Make a request for the current weather where you are born
#I was born in Colombo, Sri Lanka
response = requests.get("https://api.open-meteo.com/v1/forecast?latitude=6.9355&longitude=79.8487&current=temperature_2m&forecast_days=1")
data = response.json()
data
```

```python
#current temperature
data.keys() #exploring
current_temp = data['current']['temperature_2m']
current_temp

```

```python
#5. Print out the difference between the current temperature and how warm it feels?
response = requests.get("https://api.open-meteo.com/v1/forecast?latitude=6.9355&longitude=79.8487&current=temperature_2m,apparent_temperature")
data = response.json()
data

```

```python
#feels like
apparent_temp = data['current']['apparent_temperature']
apparent_temp
```

```python
#difference
difference= apparent_temp-current_temp
difference

print(f'It feels {difference:.2f} degrees celcius warmer')
```
